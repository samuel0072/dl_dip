```python
import cv2 as cv2
import numpy as np
from matplotlib import pyplot as plt
```

# Leitura de imagens escolhidas

```python
img1 = cv.imread('images/tanjiro.jpg',0)
img2 = cv.imread('images/02.png',0)
img3 = cv.imread('images/sami_rebeca.jpg',0)
```

## CÃ³digo de https://docs.opencv.org/4.x/de/dbc/tutorial_py_fourier_transform.html

### Imagem 1

```python
f = np.fft.fft2(img1)
fshift = np.fft.fftshift(f)
magnitude_spectrum = 20*np.log(np.abs(fshift))
```

```python
plt.subplot(121),plt.imshow(img1, cmap = 'gray')
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(magnitude_spectrum, cmap = 'gray')
plt.title('Magnitude Spectrum'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
rows, cols = img1.shape
crow, ccol = int(rows / 2), int(cols / 2)
mask = np.ones((rows, cols), np.uint8)

mask[crow-5:crow+5, ccol-5:ccol+5] = 0#mascara
```

```python
f_shift = fshift * mask
f_ishift = np.fft.ifftshift(f_shift)
img_back = np.fft.ifft2(f_ishift)
img_back = np.real(img_back)
```

```python
plt.subplot(131),plt.imshow(img1, cmap = 'gray')
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(132),plt.imshow(img_back, cmap = 'gray')
plt.title('Image after HPF'), plt.xticks([]), plt.yticks([])
plt.subplot(133),plt.imshow(img_back)
plt.title('Result in JET'), plt.xticks([]), plt.yticks([])
plt.show()
```
